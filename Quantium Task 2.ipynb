# Total Customers 

```python
total_customers = df['LYLTY_CARD_NBR'].count()
```

```python
total_customers
```

# Average number of transactions per customer

```python
df['transaction_count'] = df.groupby('LYLTY_CARD_NBR')['TXN_ID'].transform('size')
```

```python
average_transactions_per_customer = df.groupby('LYLTY_CARD_NBR')['transaction_count'].mean()
```

```python
average_transactions_per_customer
```

# Top 5 Stores with high profit
